### Histogram Equalization

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# load image
input = cv2.imread("gray.jpg")

# convert to grayscale
grayscale = cv2.cvtColor(input, cv2.COLOR_BGR2GRAY)
```

```python
# calculate histogram using Python
histogram = np.zeros((256,), dtype=np.uint32)

(r, c) = grayscale.shape

for i in range(r):
    for j in range(c):
        histogram[input[i, j]] = histogram[input[i, j]] + 1

# print(histogram)
```

```python
# calculate histogram using NumPy
histogram_np, _ = np.histogram(grayscale, bins=np.arange(257))

# print(histogram_np)
```

```python
# plot histograms
plt.subplot(121)
plt.plot(np.arange(256), histogram)
plt.subplot(122)
plt.plot(np.arange(256), histogram_np)
```

```python
# plot histogram and cumulative distribution
cdf_np = np.cumsum(histogram_np)

plt.subplot(121)
plt.plot(np.arange(256), histogram)
plt.subplot(122)
plt.plot(np.arange(256), cdf_np)
```

```python
# equalization without OpenCV
cdf = np.zeros(256, np.uint32)

cdfMin = -1

cdf[0] = histogram[0]
if histogram[0] > 0:
    cdfMin = histogram[0]
for i in range(1, 256):
    if (cdfMin == -1 and histogram[0] > 0):
        cdfMin = histogram[0]
    cdf[i] = cdf[i - 1] + histogram[i]

h = np.zeros(256, np.uint32)
for i in range(256):
    h[i] = round((float(cdf[i] - cdfMin) / float(r * c - 1)) * 255.0)

grayscale_equalized = np.zeros((r, c), np.uint8)
for i in range(r):
    for j in range(c):
        grayscale_equalized[i, j] = h[grayscale[i, j]]
```

```python
# show image
plt.imshow(grayscale_equalized, 'gray')
```

```python
# show new histogram and corresponding cdf
nh, _ = np.histogram(grayscale_equalized, bins=np.arange(257))
plt.subplot(121)
plt.plot(np.arange(256), nh)
plt.subplot(122)
plt.plot(np.arange(256), np.cumsum(nh))
```

```python
# equalization with OpenCV
grayscale_equalized_opencv = cv2.equalizeHist(grayscale)
```

```python
# compare both equalized images
plt.subplot(121)
plt.imshow(grayscale_equalized, 'gray')
plt.subplot(122)
plt.imshow(grayscale_equalized_opencv, 'gray')
```

```python
# show OpenCV equalized histogram and cdf
nh, _ = np.histogram(grayscale_equalized_opencv, bins=np.arange(257))
plt.subplot(121)
plt.plot(np.arange(256), nh)
plt.subplot(122)
plt.plot(np.arange(256), np.cumsum(nh))
```

```python
# compare histograms and cdfs
nh, _ = np.histogram(grayscale_equalized_opencv, bins=np.arange(257))
plt.subplot(222)
plt.plot(np.arange(256), nh)
plt.subplot(224)
plt.plot(np.arange(256), np.cumsum(nh))
plt.subplot(221)
plt.plot(np.arange(256), histogram)
plt.subplot(223)
plt.plot(np.arange(256), cdf_np)
```

```python
# compare images
plt.subplot(121)
plt.imshow(grayscale, 'gray')
plt.subplot(122)
plt.imshow(grayscale_equalized_opencv, 'gray')
```
